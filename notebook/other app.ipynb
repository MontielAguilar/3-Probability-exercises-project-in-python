Ejercicio 1
Se lanzan dos dados imparciales una vez y se observa la puntuación total. Utilice una simulación para encontrar la probabilidad estimada de que la puntuación total sea par o mayor que 7

Podríamos usar una ecuación matemática para resolver esto o podemos ejecutar el experimento muchas veces y luego contar la cantidad de veces en que el puntaje es par o mayor que 7.

-Ejecutar el experimento 1000 veces (tirar 2 dados 1000 veces y sumar el resultado)

-Haga un seguimiento del número de veces que la suma fue mayor que 7 o un número par

-Dividir el número del paso 2 por el número de iteraciones (1000)

```python
#1º importar bibliotecas
from random import randint                    

#2º Creamos 2 variables gobales  con int 0  que se modifican de forma local dentro del for         

contador7 = 0
contadorpar = 0
#3º Loop for donde se representaran las 1000 tiradas
for contador in range (1,1001):                             #-> for en rango de 1000
    dado1 = randint(1,6)                                    #-> 1º Dado con funcion randint (imprime aleatoriamente del 1 al 6)
    dado2 = randint(1,6)                                    #-> 2º Dado con funcion randint (imprime aleatoriamente del 1 al 6)
    dadosuma =dado1+dado2                                   #-> Variable que suma los 2 dados.
    print (f"  Has lanzado la vez {contador}. El resultado es {dadosuma}")
    contador +=1                                            #-> +1 Al contador del bucle for 
    if dadosuma == 7:                                       #-> Condicional if cuya condicion es que la suma de los dados de 7
        contador7 +=1
    elif dadosuma == 2 or 4 or 6 or 8 or 10 or 12:          #-> Condicional elif cuya condicion es que la suma de los dados de par
        contadorpar +=1

#4º División del resultado de lanzar los dados entre 1000
contador7entre1000 = contador7 / 1000                       #-> Division entre mil del resultado de veces que sale 7
contadorparentre1000 = contadorpar / 1000
#5º Print por pantalla
print ("----------------------------------")                
print (f" Ha salido el nº 7: {contador7} \n Ha salido nº par: {contadorpar}\n El % de salir el nº 7: {contador7entre1000} \n El % de salir nº par es: {contadorparentre1000}")
```

EJERCICIO 2, APARTADO 1: PROBABILIDAD DE: PROBABILIDAD DE 3 BLANCAS Y 2 ROJAS

```python
#EJERCICIO 1 PROBABILIDAD DE QUE EN LA 1 TIRADA SALGA 3 BLANCAS Y 2 ROJAS:

#----------------------------------------------------------------------------------------------------------------------------------
from math import comb                                   #-> Importamos comb de la libreria math
import numpy as np                                      #-> Importamos numpy

#--------------------------------------3 BOLAS BLANCAS-----------------------------------------------------------------------------
vecesquesacamos = 5                                     #-> veces que sacamos bolas
bolasdeexito = 3                                        # Número de bolas blancas que necesitamos sacar
p_blanca = 10 / 60                                      # probabilidad de sacar 1 bola blanca de 60 bolas

blancas3 = comb(vecesquesacamos, bolasdeexito) * p_blanca**bolasdeexito * (1 - p_blanca)**(vecesquesacamos - bolasdeexito)
        #-> formula de la binomial en funciones discretas (por ser nº enteros)
        #-> (combo n x) p(elevado x) q n-y
            #-> el combo o combinacion se hace: (formula: n! \ (n - x!x!))

print (blancas3)                                  #-> probabilidad de sacar 3 bolas blancas.

#-> RESULTADO: 0.03215020576131687

#--------------------------------------2 BOLAS ROJAS-----------------------------------------------------------------------------
# igual que antes pero cambiando bolasdeexito 2 y probabilidad de roja 20/ 60

vecesquesacamos = 5                                     #-> veces que sacamos bolas
bolasdeexito = 2                                        # Número de bolas roja que necesitamos sacar
p_roja = 20 / 60                                        # probabilidad de sacar 1 roja blanca de 60 bolas

rojas2 = comb(vecesquesacamos, bolasdeexito) * p_blanca**bolasdeexito * (1 - p_blanca)**(vecesquesacamos - bolasdeexito)
        #-> formula de la binomial en funciones discretas (por ser nº enteros)
        #-> (combo n x) p(elevado x) q n-y
            #-> el combo o combinacion se hace: (formula: n! \ (n - x!x!))

print (rojas2)                                  #-> probabilidad de sacar 3 bolas blancas.

#-> RESULTADO: 0.1607510288065844


#-------------------------------------- 3 BLANCAS Y 2 BOLAS ROJAS-----------------------------------------------------------------

blancas3rojas2 = blancas3 + rojas2
print (blancas3rojas2)

#-> RESULTADO: 0.19290123456790126
```

EJERCICIO 2, APARTADO 2: PROBABILIDAD DE 3 BLANCAS Y 2 ROJAS  O 5 MISMO COLOR SEGUN SOLUCION

```python
#EJERCICIO 2: SACAR 1000 BOLAS EXPLICADO:
#1º importa librerias
import numpy as np                                      #-> Importamos numpy
import random
d = {}                                                  #-> Variable d contiene un set vacio
for i in range(60):                                     #-> loop for en un rango de 60 veces
    if i < 10:                                          #-> condicional IF Si i (iterador es menor a 10) produce la condición
        d[i] = 'White'

    elif i > 9 and i < 30:                              #-> condicional ELIF Si i (iterador es menor 30 y 10 o mas) produce la condición
        d[i] = 'Red'

    else:                                               #-> Si no es nada de lo anterior: produce la condicion
        d[i] = 'Green'  


#Inicializar variables importantes
n_simulations = 50000                                   #-> Nº de veces que vamos a repetir la acción
part_a_total = 0                                        #-> variable que indica que sucede nuestros casos de exitos de 3 blancas y 2 rojas)
part_b_total = 0                                        #-> variable que indica que sucede nuestros casos de exitos de 5 iguales)

for i in range(n_simulations):                          #-> loop for en rango de la variable (50.000)

    list = []                                           #-> Lista vacia que aprende del loop for que simula sacar 5 bolas
    
    for i in range(5):                                  #-> loop for que simula sacar 5 bolas
        list.append(d[random.randint(0,59)])            #-> La lista vacia aprende del resultado de:
                                                            #-> d (set que recoge las bolas blancas, rojas o verdes)
                                                            #-> y de ese set sacamos 1 bola aleatoria (por ello el rango de 0, 59)
                                                        
        
#Convertir en array
    list = np.array(list)                               #-> simplemente aplicamos la funcion array de numpy sobre la lista

#sumatorio de blancos, rojos y verdes
    white = sum(list == 'White')                        #->creamos la funcion blanca que es = suma cuando en la lista salga white
    red = sum(list == 'Red')                            #->creamos la funcion roja que es = suma cuando en la lista salga red
    green = sum(list == 'Green')                        #->creamos la funcion verde que es = suma cuando en la lista salga green

#Keep track if the combination met the criteria
    if white == 3 and red == 2:                         #-> Condicional if que dice que si white y red se cumplen 
        part_a_total += 1                               #-> le sumamos 1 a la variable part_a_total que inicialmente es cero
    
    if red == 5 or white == 5 or green ==5:             #-> Condicional if que dice que si white o red o green son 5: 
        part_b_total +=1                                #-> le sumamos 1 a la variable part_a_total que inicialmente es cero


print('The probability of 3 white and 2 red is:', round(part_a_total/n_simulations*100,2),'%')
print('The probability of all the same color is:', round(part_b_total/n_simulations*100,2),'%') 

#-> aplicamos al formula de la probabilidad entre casos de existo / casos posibles *100
    #-> El ,2 indica que esta puesto con el limite de 2 decimales.
```

EJERCICIO 2, APARTADO 2: PROBABILIDAD DE 3 BLANCAS Y 2 ROJAS  O 5 MISMO COLOR.

```python
#EJERCICIO 2: OTRA FORMA DE EJECUTAR CON PEQUEÑAS MODIFICACIONES
import random

#Variables: 
bolastotal = 60                                     #-> total de numero de bolas
blancas = 10                                        #-> nº bolas blancas
rojas = 20                                          #-> nº bolas rojas
verdes = 30                                         #-> nº bolas verdes
num_simulaciones = 1000                             #-> nº de veces que ejecutamos sacar 5 bolas
cont_3_blancos_2_rojos = 0                          #-> Contador de 3 blancas y 2 rojas
cont_5_mismo_color = 0

# Realizar las simulaciones
for _ in range(num_simulaciones):                   #-> Ejecutamos la acción. Loop for que repite mil veces lo que contiene
    cont_blancas = 0
    cont_rojas = 0
    for _ in range(5):                              #-> loop en rango 5 dentro de loop en rango 1000 (es la accion de sacar 5 bolas con reemplazo)
        seleccionar_bola = random.randint(1, bolastotal)
                                                    #-> dentro de este for seleccionamos las bolas de forma aleatoria segun random.randint
        if seleccionar_bola <= blancas:             #-> condicional if: si es blanca suammos a contador
            cont_blancas += 1                       #-> +1 contador
        elif seleccionar_bola <= blancas + rojas:   #-> condicional if: si es roja suammos a contador
            cont_rojas += 1                         #-> +1 contador
    
    # Verificar los eventos de interés
    if cont_blancas == 3 and cont_rojas == 2:       #-> operación de suma bolas 3 blancas y 2 rojas
        cont_3_blancos_2_rojos += 1                 #-> +1
    if cont_blancas == 5 or cont_rojas == 5:        #-> operacion de suma de todas las bolas mismo color
        cont_5_mismo_color += 1                     #-> +1

# Calcular las probabilidades
prob_3_blancos_2_rojos = cont_3_blancos_2_rojos / num_simulaciones  #-> formula de la probabilidad
prob_5_mismo_color = cont_5_mismo_color / num_simulaciones          #-> formula de la probabilidad

# Imprimir resultados
print(f"Probabilidad: 3 blancas y 2 rojas: {prob_3_blancos_2_rojos}")
print(f"Probabilidad:5 bolas sean del mismo color: {prob_5_mismo_color}")
```
